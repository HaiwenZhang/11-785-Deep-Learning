```python
import numpy as np
import time
from scipy.special import expit
from skeleton import *
```

```python
#PROBLEM ZERO TEST
X,Y = load_data('digitstrain.txt')
```

```python
def update_weights_perceptron_vec(X, Y, weights, bias, lr):
    
    """
    update the weight and bias of a NN with a single sigmoid activation layer
    and softmax output.
    Divergence measure: cross entropy
    """
    
    updated_weights = [] 
    updated_bias = []
    
    gradW1 = np.zeros((784,10))
    gradb1 = np.zeros(10)
    
    T = len(Y)#batch size
    
    Z1 = np.dot(X,weights[0]) + bias[0]
    Y1 = expit(Z1)
    
    exp_Y1 = np.exp(Y1)
    output = exp_Y1/np.sum(exp_Y1,axis=1,keepdims=True)
    
    #output = softmax(Y1)#apply softmax on last layer
    loss   = np.sum(-np.log(output[[range(T),Y]]))/T

    output[[range(T),Y]] -= 1
    gradDivY1 = output
    
    sigmoidJacobian = Y1*(1-Y1)
    gradDivZ1 = sigmoidJacobian*gradDivY1#
    gradb1 = np.sum(gradDivZ1,axis=0,keepdims=False)
    gradW1 = np.dot(np.array(X).T,gradDivZ1)
    updated_weights.append(weights[0]-lr*(1/T)*gradW1)
    updated_bias.append(bias[0]-lr*(1/T)*gradb1)
    

    return updated_weights, updated_bias, loss
```

```python
#PROBLEM ONE TEST
weights = [np.arange(-0.01,0.01,0.02/(10*784)).reshape(784,10)]
bias    = [np.zeros(10)]
lr      = 0.001
#[weights,bias] = update_weights_perceptron_vec(X, Y, weights, bias, lr)
t = time.time()
loss_history = []
for i in range(100):
    [weights,bias,loss] = update_weights_perceptron_vec(X, Y, weights, bias, lr)
    if i%10:
        print('loss:', loss)
    loss_history.append(loss)
print('it took',int(time.time()-t),' seconds')
```
